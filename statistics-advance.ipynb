```python
# Q1. What is a random variable in probability theory?
# Ans. 
'''
A random variable is a numerical outcome of a random phenomenon. It assigns a real number to each possible outcome in a sample space. 
Example: Tossing a coin—define X = 1 for heads, X = 0 for tails. Here, X is a random variable.
'''
```

```python
#
```

```python
# Q2. What are the types of random variables?
# Ans
'''
    Random variables are of two main types:
    - Discrete: Takes specific countable values (e.g., number of students).
    - Continuous: Takes any value in a range (e.g., height, weight).
    Example: X = number of cars per minute is discrete, Y = time taken by a car to cross a toll is continuous.
'''
```

```python
#
```

```python
# Q3. What is the difference between discrete and continuous distributions?
# Ans. 
'''
    Discrete Distribution: Defined for countable outcomes.
    Continuous Distribution: Defined over a continuous range of values.
    Example: Binomial (discrete), Normal (continuous).
'''
```

```python
#
```

```python
# Q4. What are probability distribution functions (PDF)?
# Ans.
'''
    A Probability Distribution Function (PDF) describes the likelihood of a random variable taking specific values.
    For discrete variables: Called Probability Mass Function (PMF).

    For continuous variables: Called PDF.
    Example in Python: scipy.stats.norm.pdf(x) gives the PDF of normal distribution at x.
'''
```

```python
#
```

```python
# Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
# Ans.
'''
    PDF gives the probability density at a specific point.
    CDF gives the probability that a variable is less than or equal to a value.
    Example: scipy.stats.norm.cdf(1.96) gives area under the normal curve to the left of 1.96.
'''
```

```python
#
```

```python
# Q6. What is a discrete uniform distribution?
# Ans.
'''
    A discrete uniform distribution assigns equal probability to each of the n outcomes.
    Example: Rolling a fair die. Each face (1–6) has probability 1/6.
'''
```

```python
#
```

```python
# Q7. What are the key properties of a Bernoulli distribution?
# Ans.
'''
    Two outcomes: success (1) and failure (0)
    Probability of success = p, failure = 1-p
    Mean = p, Variance = p(1-p)
    Example: Tossing a biased coin where p = 0.7 for heads.
'''
```

```python
#
```

```python
# Q8. What is the binomial distribution, and how is it used in probability?
# Ans.
'''
    The binomial distribution models the number of successes in n independent Bernoulli trials.
    Formula: P(X = k) = C(n, k) * p^k * (1-p)^(n-k)
    Example: Probability of 3 heads in 5 tosses of a biased coin.
'''
```

```python
#
```

```python
# Q9. What is the Poisson distribution and where is it applied?
# Ans.
'''
    The Poisson distribution models the number of events in a fixed interval when they occur independently and at a constant rate λ.
    Example: Number of customer arrivals at a shop per hour.
'''
```

```python
#
```

```python
# Q10. What is a continuous uniform distribution?
# Ans. 
'''
    A continuous uniform distribution has constant probability over a continuous range [a, b].
    PDF = 1 / (b - a)
    Example: Randomly choosing a time between 2:00 and 3:00 PM.
'''
```

```python
#
```

```python
# Q11. What are the characteristics of a normal distribution?
# Ans.
'''
    - Bell-shaped and symmetric
    - Mean = Median = Mode
    - Defined by mean (μ) and standard deviation (σ)
      Example: Heights of students often follow a normal distribution.
'''
```

```python
#
```

```python
# Q12. What is the standard normal distribution, and why is it important?
# Ans.
'''
    A standard normal distribution is a normal distribution with mean = 0 and standard deviation = 1.
    It is used for standardizing data and calculating Z-scores.
    Example: Z-tables are based on the standard normal distribution.
'''
```

```python
#
```

```python
# Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
# Ans.
'''
    The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population's distribution.
    Importance: Enables use of normal distribution in hypothesis testing and confidence intervals.
'''
```

```python
#
```

```python
# Q14. How does the Central Limit Theorem relate to the normal distribution?
# Ans.
'''
    CLT explains why sample means are normally distributed, even when population distribution is not.
    Example: Sampling average income from a skewed population still results in a normal-shaped mean distribution if sample size is large (n > 30).
'''
```

```python
#
```

```python
# Q15.What is the application of Z statistics in hypothesis testing?
# Ans.
'''
    Z-statistics helps determine how far a sample statistic is from the population mean under the null hypothesis.
    Used in large-sample tests for means and proportions.
    Example: Testing if a new drug improves recovery time compared to an existing one.
'''
```

```python
#
```

```python
# Q16. How do you calculate a Z-score, and what does it represent?
# Ans.
'''
    Z = (X - μ) / σ
    Represents how many standard deviations X is from the mean.
    Example: If X = 80, μ = 70, σ = 5 → Z = 2 (i.e., 2 SD above the mean)
'''
```

```python
#
```

```python
# Q17. What are point estimates and interval estimates in statistics?
# Ans.
'''
    Point Estimate: A single value (like sample mean) used to estimate a population parameter.
    Interval Estimate: A range (like confidence interval) that likely contains the parameter.
    Example: Point estimate of average age = 25; 95% CI = [23, 27]
'''
```

```python
#
```

```python
# Q18. What is the significance of confidence intervals in statistical analysis?
# Ans.
'''
    A confidence interval (CI) provides a range where the true population parameter is likely to fall with a certain level of confidence (e.g., 95%).
    Example: A CI for average test score means we’re 95% sure the true mean is within that range.
'''
```

```python
#
```

```python
# Q19. What is the relationship between a Z-score and a confidence interval?
# Ans.
'''
    Z-scores determine the margin of error in confidence intervals.
    For 95% CI, Z = 1.96
    CI formula: mean ± Z * (σ / √n)
'''
```

```python
#
```

```python
# Q20. How are Z-scores used to compare different distributions?
# Ans.
'''
    Z-scores standardize values from different distributions, making them comparable.
    Example: Comparing exam scores from different subjects with different means and variances.
'''
```

```python
#
```

```python
# Q21. What are the assumptions for applying the Central Limit Theorem?
# Ans.
'''
    - Independent samples
    - Identically distributed
    - Sample size should be sufficiently large (n ≥ 30)
      CLT applies even if the population distribution is not normal.
'''
```

```python
#
```

```python
# Q22. What is the concept of expected value in a probability distribution?
# Ans.
'''
    The expected value (E[X]) is the theoretical mean of a random variable.
    Formula: E[X] = Σ (x * P(x))
    Example: For a die: E[X] = (1+2+3+4+5+6)/6 = 3.5
'''
```

```python
#
```

```python
# Q23.How does a probability distribution relate to the expected outcome of a random variable?
# Ans.
'''
    The probability distribution defines how likely each outcome is.
    The expected outcome (mean) is a weighted average of those outcomes.
    Example: In betting, the expected value tells you your average win/loss over many games.
'''
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
import numpy as np
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    return x, bernoulli.pmf(x, p)

x, pdf = bernoulli_pdf(0.6)
print("PDF:", list(zip(x, pdf)))
```

```python
# 12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)?
import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    y = (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)
    plt.plot(x, y)
    plt.title("Standard Normal Distribution (mean=0, std=1)")
    plt.xlabel("x")
    plt.ylabel("PDF")
    plt.grid()
    plt.show()

plot_standard_normal()
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset?
from scipy.stats import norm
import numpy as np

# Sample data
sample_mean = 105
population_mean = 100
std_dev = 15
n = 30
z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z)
print(f"Z = {z}, p-value = {p_value}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result?
import scipy.stats as stats
import numpy as np

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
sem = stats.sem(data)
conf_int = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)
print(f"95% Confidence Interval: {conf_int}")
```
