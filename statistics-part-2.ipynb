```python
# Q1. What is hypothesis testing in statistics?
# Ans.
'''
Hypothesis testing is a method used to decide whether there is enough statistical evidence to support a specific belief (hypothesis) about a population parameter.
Example: Testing if a new marketing strategy increases sales compared to the old one.
'''
```

```python
#
```

```python
# Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
# Ans.
'''
Null Hypothesis (H₀): States that there is no effect or difference.

Alternative Hypothesis (H₁ or Ha): States that there is an effect or difference.
Example:
    H₀: μ = 50 (no change),
    H₁: μ ≠ 50 (some change)
'''
```

```python
#
```

```python
# Q3. What is the significance level in hypothesis testing, and why is it important?
# Ans.
'''
The significance level (α) is the threshold for rejecting the null hypothesis. Common values are 0.05 or 0.01.
It defines the probability of making a Type I error.
Example: 
    α = 0.05 means a 5% risk of wrongly rejecting H₀.
'''
```

```python
#
```

```python
# Q4. What does a P-value represent in hypothesis testing?
# Ans.
'''
The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming H₀ is true.
Lower P-values suggest stronger evidence against H₀.
'''
```

```python
#
```

```python
# Q5. How do you interpret the P-value in hypothesis testing?
# Ans.
'''
If P ≤ α, reject H₀ → statistically significant

If P > α, fail to reject H₀ → not statistically significant
Example: 
    P = 0.03 and α = 0.05 → Reject H₀
'''
```

```python
#
```

```python
# Q6. What are Type 1 and Type 2 errors in hypothesis testing?
# Ans.
'''
Type I Error: Rejecting H₀ when it is true (false positive)

Type II Error: Failing to reject H₀ when it is false (false negative)
Example: 
    Declaring a drug effective when it is not → Type I
'''
```

```python
#
```

```python
# Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
# Ans.
'''
One-tailed test: Checks deviation in one direction (greater/less)

Two-tailed test: Checks deviation in both directions
Example:
    One-tailed → H₁: μ > 50
    Two-tailed → H₁: μ != 50
'''
```

```python
#
```

```python
# Q8. What is the Z-test, and when is it used in hypothesis testing?
# Ans.
'''
A Z-test is used when the population standard deviation is known and the sample size is large (n > 30).
It tests means, proportions, or differences.
Example: 
    Comparing sample mean to population mean using Z.
'''
```

```python
#
```

```python
# Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
# Ans.

# Formula:
# Z = (x̄ - μ) / (σ / √n)
#  where:
# x̄ = sample mean
# μ  = population mean
# σ  = population standard deviation
# n  = sample size

# It measures how many standard deviations a sample mean is from the population mean under H₀.
```

```python
#
```

```python
# Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
# Ans.
'''
The T-distribution is used when the population standard deviation is unknown and the sample size is small (n < 30).
It has heavier tails than the normal distribution.
Example: 
    Small sample t-test for student scores.
'''
```

```python
#
```

```python
# Q11. What is the difference between a Z-test and a T-test?
# Ans.
'''
Z-test: Known σ, large n

T-test: Unknown σ, small n
Both compare means but use different distributions.
'''
```

```python
# Q12. What is the T-test, and how is it used in hypothesis testing?
# Ans.
'''
A T-test evaluates if the means of two groups are significantly different.
Types: One-sample, two-sample, paired.
Example: 
    Comparing average test scores of two classrooms.
'''
```

```python
# Q13. What is the relationship between Z-test and T-test in hypothesis testing?
# Ans.
'''
Both test hypotheses about means. 
The T-test approximates the Z-test as sample size increases due to the Central Limit Theorem. 
They differ in standard deviation assumptions.
'''
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?
import numpy as np
from scipy.stats import norm

sample = np.array([102, 98, 100, 101, 99])
population_mean = 100
population_std = 2

sample_mean = np.mean(sample)
n = len(sample)
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

sample = np.random.normal(105, 10, 40)
z, p = one_sample_z_test(sample, 100, 10)
print(f"Z = {z:.2f}, P-value = {p:.4f}")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z_stat = 2.1
alpha = 0.05
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Z-distribution')
z_crit = norm.ppf(1 - alpha / 2)
plt.fill_between(x, y, where=(x <= -z_crit) | (x >= z_crit), color='red', alpha=0.3, label='Rejection region')
plt.axvline(z_stat, color='black', linestyle='--', label='Z-statistic')
plt.legend()
plt.title('Two-tailed Z-test')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu0 = 0  # Null mean
mu1 = 1  # Alternative mean
sigma = 1
alpha = 0.05
x = np.linspace(-4, 4, 1000)
z_crit = norm.ppf(1 - alpha)

null_dist = norm.pdf(x, mu0, sigma)
alt_dist = norm.pdf(x, mu1, sigma)

plt.plot(x, null_dist, label='H0: μ=0')
plt.plot(x, alt_dist, label='H1: μ=1')
plt.axvline(z_crit, color='black', linestyle='--', label='Critical Value')
plt.fill_between(x, null_dist, where=(x >= z_crit), color='red', alpha=0.3, label='Type I Error')
plt.fill_between(x, alt_dist, where=(x < z_crit), color='blue', alpha=0.3, label='Type II Error')
plt.legend()
plt.title('Type I and Type II Errors')
plt.grid()
plt.show()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance?
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = np.std(data, ddof=1) / np.sqrt(len(data))
    margin = norm.ppf((1 + confidence) / 2) * se
    return mean - margin, mean + margin

data = np.random.normal(50, 5, 30)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example: Disease test
p_disease = 0.01
p_positive_given_disease = 0.99
p_positive_given_no_disease = 0.05

posterior = bayes_theorem(p_disease, p_positive_given_disease, p_positive_given_no_disease)
print(f"Probability of having disease given a positive test: {posterior:.4f}")
```
