**Question 1:** What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

**Ans:** Anomaly Detection is the process of identifying data points, events, or observations that do not conform to the expected pattern of a given dataset. These non-conforming instances are often referred to as **anomalies**, **outliers**, **exceptions**, or **novelties**. The primary goal is to distinguish these rare occurrences from the normal, majority data. üßê

There are three main types of anomalies:

1.  **Point Anomaly**: A single data instance is considered anomalous if it deviates significantly from the rest of the data. This is the simplest and most common type of anomaly.
    * **Example**: In a dataset of human body temperatures, a reading of 105¬∞F would be a point anomaly, as most readings would cluster around 98.6¬∞F. Similarly, a credit card transaction of $5,000 for a user whose average transaction is $40 is a clear point anomaly.

2.  **Contextual Anomaly**: A data instance is anomalous within a specific context, but not otherwise. The "context" is usually temporal (time-based) or spatial.
    * **Example**: Spending $300 on toys is normal in December (the context is the holiday season). However, the exact same purchase in May would be unusual and thus a contextual anomaly. The data point itself is normal, but its occurrence in the wrong context is not.

3.  **Collective Anomaly**: A collection of related data instances is anomalous as a group, even though the individual data points within that group may not be anomalies by themselves.
    * **Example**: In an Electrocardiogram (ECG) recording of a human heartbeat, a prolonged flat-line signal is a dangerous collective anomaly (indicating cardiac arrest). While each individual data point in the flat-line is within a normal range, their collective occurrence and sequence are highly abnormal.

**Ans:** Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are all powerful algorithms for detecting anomalies, but they operate on different principles and are suited for different scenarios.

* **Local Outlier Factor (LOF)**: Analyzing manufacturing sensor data where one machine might operate at a different (but normal) density of readings than another. LOF can spot an anomaly within one machine's context without flagging it as an outlier relative to the whole system.


**Question 3:** What are the key components of a Time Series? Explain each with one example.

**Ans:** A time series is a sequence of data points recorded at consistent time intervals. Any time series can be broken down into four key components that, when combined, represent the original data.

1.  **Trend ($T_t$)**: This is the **long-term direction** or underlying movement of the data over time. A trend can be increasing, decreasing, or stable.
    * **Example**: The steady increase in the global population over the last century shows a clear upward trend. üìà

2.  **Seasonality ($S_t$)**: This is a **repeating, predictable pattern** that occurs at fixed and known intervals (e.g., daily, weekly, monthly, yearly).
    * **Example**: Retail sales data showing a large spike every November and December due to holiday shopping demonstrates strong yearly seasonality. üéÅ

3.  **Cyclical Component ($C_t$)**: This refers to patterns that repeat over **non-fixed, longer periods** of time. Cycles are often related to broader economic or business conditions and are harder to predict than seasonality.
    * **Example**: The real estate market often exhibits cyclical behavior, with periods of growth ("boom") and decline ("bust") that can span several years. üè°

4.  **Irregularity or Residual ($R_t$)**: This component represents the **random, unpredictable fluctuations** or "noise" in the data that cannot be explained by the other three components.
    * **Example**: A sudden, one-day surge in a company's stock price due to an unexpected news event is an irregular component. ‚ö°

These components can be combined in an **additive model** ($Y_t = T_t + S_t + C_t + R_t$) or a **multiplicative model** ($Y_t = T_t \times S_t \times C_t \times R_t$).

**Question 4:** Define Stationarity in time series. How can you test and transform a non-stationary series into a stationary one?

**Ans:** Stationarity is a fundamental concept in time series analysis. A time series is considered **stationary** if its statistical properties‚Äîspecifically its **mean, variance, and autocorrelation**‚Äîare all constant over time. In simpler terms, a stationary series does not have a trend or seasonal patterns, and its fluctuations are consistent across time.

Stationarity is important because most time series forecasting models are designed to work on stationary data.

**How to Test for Stationarity**
1.  **Visual Inspection**: Plot the time series. If you can clearly see a trend (long-term increase/decrease) or obvious seasonality, the series is likely non-stationary.
2.  **Augmented Dickey-Fuller (ADF) Test**: This is a formal statistical test for stationarity.
    * **Null Hypothesis ($H_0$)**: The time series is non-stationary.
    * **Alternative Hypothesis ($H_1$)**: The time series is stationary.
    * If the test's **p-value is below a significance level** (e.g., 0.05), you reject the null hypothesis and conclude that the series is stationary.

**How to Transform a Non-Stationary Series into a Stationary One**
1.  **Differencing**: This is the most common method to remove trends and seasonality. You transform the series by subtracting the previous observation from the current observation ($Y'_t = Y_t - Y_{t-1}$).
2.  **Log Transformation**: If the series shows non-constant variance (i.e., the fluctuations get bigger over time), applying a logarithmic transformation ($log(Y_t)$) can help stabilize it.
3.  **Detrending**: You can fit a regression line to the data to model the trend and then subtract this trend from the original series.

**Question 5:** Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

**Ans:** These models are all part of the popular ARIMA family used for time series forecasting, with each building upon the last.

* **AR (Autoregressive)**: An AR model predicts future values based on a linear combination of its **own past values**. The term `p` represents the number of past lags used.
    * **Application**: Used when the current value in a series is strongly correlated with its previous values.

* **MA (Moving Average)**: An MA model predicts future values based on a linear combination of **past forecast errors**. The term `q` represents the number of past errors used.
    * **Application**: Used to model shocks or random events whose effects are felt over time.

* **ARIMA (Autoregressive Integrated Moving Average)**: This model combines AR and MA components and adds **differencing** (the 'I' for Integrated) to handle non-stationary data.
    * **Application**: The workhorse of time series forecasting for non-seasonal data with a trend.

* **SARIMA (Seasonal ARIMA)**: This is an extension of ARIMA that directly incorporates **seasonal components**.
    * **Application**: Essential for forecasting data with clear seasonality, like monthly retail sales or airline passenger numbers.

* **SARIMAX (SARIMA with eXogenous Variables)**: This is the most versatile model in the family. It's a SARIMA model that also includes **external predictor variables** (exogenous variables).
    * **Application**: Perfect for situations where external factors influence the time series. For example, forecasting electricity demand (`Y_t`) using weather temperature (`X_t`) as an exogenous variable.

**Question 6:** Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.

**Ans:** This code loads the classic "AirPassengers" dataset, which contains monthly international airline passenger numbers from 1949 to 1960. It then uses the `seasonal_decompose` function from the `statsmodels` library to break the time series down into its three core components: the long-term **Trend**, the yearly **Seasonality**, and the leftover **Residuals**.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 1. Load the dataset
url = 'https://raw.githubusercontent.com/barman-ankita/PWskills/refs/heads/main/AirPassengers.csv'
df = pd.read_csv(url, header=0, index_col=0, parse_dates=True)
df.index.name = 'Month'

# Clean column names (remove '#' and spaces)
df.columns = df.columns.str.replace('#', '').str.strip()

print(df.head())
print(df.columns)  # should now show 'Passengers'

# 2. Decompose the time series
decomposition = seasonal_decompose(df['Passengers'], model='multiplicative', period=12)

# 3. Plot the original series and the decomposed components
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.suptitle('Time Series Decomposition of Air Passengers Data', y=1.02)
plt.tight_layout()
plt.show()
```

**Question 7:** Apply Isolation Forest on a numerical dataset to detect anomalies. Visualize the anomalies on a 2D scatter plot.

**Ans:** This example demonstrates Isolation Forest on a synthetic 2D dataset. The code creates a large cluster of "normal" data points and a few "anomalous" points far away. The final scatter plot clearly distinguishes the normal points from the anomalies, which are circled in red.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

# 1. Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=1.0, random_state=42)
outliers = np.random.uniform(low=-8, high=8, size=(20, 2))
X = np.vstack([X, outliers])

# 2. Apply Isolation Forest
iso_forest = IsolationForest(n_estimators=100, contamination=0.06, random_state=42)
y_pred = iso_forest.fit_predict(X)
anomalies = X[y_pred == -1]
inliers = X[y_pred == 1]

# 3. Visualize the anomalies
plt.figure(figsize=(8, 6))
plt.scatter(inliers[:, 0], inliers[:, 1], c='skyblue', label='Normal Data')
plt.scatter(anomalies[:, 0], anomalies[:, 1],
            edgecolor='red', facecolor='none', s=150, label='Anomalies')
plt.title('Anomaly Detection using Isolation Forest')
plt.legend()
plt.grid(True)
plt.show()
```

The output is a scatter plot. The majority of the points form a dense blue cloud. A few scattered points, far from this central cloud, are highlighted with large red circles, indicating they are anomalies.

**Question 8:** Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

**Ans:** This code trains a SARIMA model on the AirPassengers dataset, which is ideal for data with both a trend and seasonality. The model is trained on the full dataset and then used to forecast the next 12 months.

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# 1. Load and prepare the data
url = "https://raw.githubusercontent.com/barman-ankita/PWskills/refs/heads/main/AirPassengers.csv"
df = pd.read_csv(url, header=0, index_col=0, parse_dates=True)
df.index.name = 'Month'

# 2. Train the SARIMA model
model = sm.tsa.statespace.SARIMAX(df['Passengers'],
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12))
results = model.fit(disp=False)

# 3. Forecast the next 12 months
forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()

# 4. Visualize the forecast
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Passengers'], label='Observed')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, color='red', label='Forecast')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:, 0],
                 forecast_ci.iloc[:, 1], color='pink', alpha=0.5)
plt.title('Air Passengers Forecast with SARIMA')
plt.legend()
plt.grid(True)
plt.show()
```

The output is a line graph showing the historical data in blue and the 12-month forecast in red. The model correctly captures the upward trend and seasonal pattern. A shaded pink area represents the 95% confidence interval.

**Question 9:** Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them.

**Ans:** This example showcases the strength of Local Outlier Factor (LOF) in identifying anomalies in datasets with varying densities. The code generates two distinct clusters (one dense, one sparse) and a few outliers. LOF successfully identifies the outliers by comparing the local density of each point to that of its neighbors.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# 1. Generate synthetic data with varying densities
np.random.seed(42)
X1 = np.random.randn(100, 2) + np.array([2, 2])
X2 = np.random.randn(50, 2) * 2.5 + np.array([-5, -5])
outliers = np.random.uniform(low=-10, high=10, size=(10, 2))
X = np.vstack([X1, X2, outliers])

# 2. Apply Local Outlier Factor (LOF)
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
y_pred = lof.fit_predict(X)
anomalies = X[y_pred == -1]
inliers = X[y_pred == 1]

# 3. Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(inliers[:, 0], inliers[:, 1], c='skyblue', label='Normal Data')
plt.scatter(anomalies[:, 0], anomalies[:, 1],
            edgecolor='red', facecolor='none', s=150, label='Anomalies')
plt.title('Anomaly Detection using Local Outlier Factor (LOF)')
plt.legend()
plt.grid(True)
plt.show()
```

The scatter plot displays two blue clouds of points (one tight, one spread out). A few individual points, located far from either cluster, are circled in red, demonstrating LOF's ability to identify outliers in varying density contexts.

- For short-term demand forecasting, I would use a SARIMAX model.

  - Why?: The S (Seasonal) component can capture daily and weekly energy usage patterns. The X (eXogenous) component is critical, as it allows the inclusion of powerful external predictors like weather conditions, time of day, and day of the week.
3. How would you validate and monitor the performance over time?
  - Initial Validation: Use time series cross-validation on historical data, evaluating with metrics like Mean Absolute Error (MAE).

  - Ongoing Monitoring: Deploy a live dashboard to track the forecast error (forecast vs. actual) in real-time. Set up automated alerts to trigger model retraining if the error exceeds a predefined threshold.
4. How would this solution help business decisions or operations?
Cost Savings: Accurate demand forecasting allows the company to optimize power generation, avoiding overproduction or having to buy expensive energy on the spot market.

  - Grid Stability: Real-time anomaly detection acts as an early warning system for equipment failure or grid instability, allowing operators to take proactive measures to prevent blackouts.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# 1. Simulate placeholder power grid data
timestamps = pd.to_datetime(pd.date_range('2025-01-01', periods=1000, freq='15min'))
data = {
    'energy_usage': 100 + np.arange(1000) * 0.1 + np.sin(np.arange(1000) / 96 * 2 * np.pi) * 20,
    'temperature': 15 + np.sin(np.arange(1000) / 96 * 2 * np.pi) * 10
}
df = pd.DataFrame(data, index=timestamps)
y = df['energy_usage']
X = df[['temperature']]

# 2. Define and train the SARIMAX model
sarimax_model = sm.tsa.statespace.SARIMAX(y, exog=X,
                                          order=(1, 1, 1),
                                          seasonal_order=(1, 1, 1, 96))
results = sarimax_model.fit(disp=False)
print("SARIMAX Model Summary:")
print(results.summary())
```
